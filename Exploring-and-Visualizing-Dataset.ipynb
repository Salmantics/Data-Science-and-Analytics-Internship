### Step 1: Import Required Libraries
In this step, we import the core libraries needed for data loading, summarization, and visualization:

- **pandas** → data handling  
- **seaborn** → provides the built-in Iris dataset  
- **matplotlib & seaborn** → for visualizations  

These libraries will help us read the dataset, explore it, and create visual plots.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# For better plot style
sns.set(style="whitegrid")
```

### Step 2: Load the Iris Dataset
We load the famous Iris dataset using seaborn's built-in function.
The dataset contains:

- Sepal Length  
- Sepal Width  
- Petal Length  
- Petal Width  
- Species (class)

After loading, we preview the first few rows to understand the dataset structure.

```python
# Load Iris dataset
data = sns.load_dataset("iris")

# Display first few rows
data.head()
```

### Step 3: Display Dataset Structure
We inspect the dataset using:

- `.shape` → number of rows & columns  
- `.columns` → feature names  
- `.head()` → preview sample data  

This helps us understand what data we are working with.

```python
print("Dataset Shape:", data.shape)
print("\nColumn Names:", data.columns.tolist())

# Display dataset preview again
data.head()
```

### Step 4: Scatter Plot — Exploring Relationships
A scatter plot helps us observe relationships between two numeric variables.

Here, we visualize:
**Petal Length vs Petal Width**, colored by species.

This helps identify clusters and separation between classes.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x="petal_length", y="petal_width", hue="species")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```

### Step 5: Histogram — Understanding Data Distribution
A histogram helps us analyze how a variable's values are distributed.

Here, we plot a histogram of **sepal_length** to observe:

- Center of distribution  
- Spread  
- Shape (normal, skewed, etc.)  


```python
plt.figure(figsize=(8,5))
sns.histplot(data["sepal_length"], kde=True)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
###  Step 6: Box Plot — Detecting Outliers & Spread
A box plot helps identify:

- Median  
- Quartiles  
- Outliers  
- Distribution spread  

Here, we visualize **sepal_width** across different species.
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=data, x="species", y="sepal_width")
plt.title("Box Plot of Sepal Width by Species")
plt.show()
```
